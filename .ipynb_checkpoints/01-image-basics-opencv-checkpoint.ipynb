# 3. Drawing on Images using OpenCV

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
blank_img.shape
```

```python
# total black
plt.imshow(blank_img)
plt.show()
```

```python
cv2.rectangle(blank_img, pt1=(384, 10), pt2=(500, 150), color=(0, 255, 0), thickness=10)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.rectangle(blank_img, pt1=(200, 200), pt2=(300, 300), color=(0, 0, 255), thickness=10)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.circle(img=blank_img, center=(250, 250), radius=150, color=(255, 0, 0), thickness=8)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.circle(img=blank_img, center=(450, 450), radius=25, color=(255, 0, 0), thickness=-1)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.line(blank_img, pt1=(0, 0), pt2=(512, 512), color=(102, 255, 255), thickness=5)
plt.imshow(blank_img)
plt.show()
```

```python
# add text
# we have to get font which are builtin to opencv
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(blank_img, text="Hello", org=(10, 500), fontFace=font, fontScale=4, color=(255, 255, 255), thickness=3, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
plt.show()
```

```python
# drwaing polygon
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int32)
plt.imshow(blank_img)
plt.show()
```

```python
vertices = np.array([ (100, 300), (200, 200), (400, 300), (200, 400) ], dtype=np.int32)
```

```python
vertices.shape
```

```python
# we need to cast them into 3d to get one color channel
pts = vertices.reshape((-1, 1, 2))
```

```python
pts.shape
```

```python
cv2.polylines(blank_img, [pts], isClosed=True, color=(255, 0, 0), thickness=5)
plt.imshow(blank_img)
plt.show()
```

# 4. Drawing on Images with Mouse

```python
import cv2
import numpy as np


#################
# FUNCTION
################
def draw_circle(event, x, y, flags, param):
    # left button clicked
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img, (x, y), 50, (0, 255, 0), -1)
    # right button clicked
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img, (x, y), 50, (255, 0, 0), -1)

cv2.namedWindow(winname="my_drawing")

cv2.setMouseCallback("my_drawing", draw_circle)



################################
# SHOWING IMAGES WITH OPENCV
################################

img = np.zeros((512, 512, 3), np.int8)

while True:
    cv2.imshow('my_drawing', img)
    # escape key
    if cv2.waitKey(20) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()
```

# 5. Creating dynamic size rectangle

```python
# VARIABLES
# false if mouse button is pressed and true if mouse button is released
drawing = False
ix = -1
iy = -1



# FUNCTION
def draw_rectangle(event, x, y, flags, params):
    global ix, iy, drawing
    
    if event == cv2.EVENT_LBUTTONDOWN:
        
        drawing = True
        ix, iy = x, y
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            cv2.rectangle(img, (ix, iy), (x, y), color=(0, 255, 0), thickness=2)
            
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(img, (ix, iy), (x, y), color=(0, 255, 0), thickness=2)


# SHOWING THE IMAGE
img = np.zeros((512, 512, 3))
cv2.namedWindow(winname="my_drawing")
cv2.setMouseCallback("my_drawing", draw_rectangle)


while True:
    cv2.imshow("my_drawing", img)
    
    if cv2.waitKey(1) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()
```
